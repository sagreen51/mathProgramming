b. (4pts) On the interval $[0,\pi]$, plot the semi-log error in your approximation for tolerance values $10^{-4}$, $10^{-6}$, and $10^{-8}$.

```python
def exp_func(x,tol):
    tot = 1. 
    aj = 1. 
    j = 1 
    while np.abs(aj)>=tol:
        aj *= x/j
        tot += aj
        j+=1 
    return tot 
```

```python
xvals = np.linspace(0.,3.14,int(1e3))
yvalst = np.exp(xvals)
yvals1 = [exp_func(xval,1e-5) for xval in xvals]
yvals2 = [exp_func(xval,1e-7) for xval in xvals]
yvals3 = [exp_func(xval,1e-9) for xval in xvals]

plt.plot(xvals,np.abs(yvalst-yvals1),color='r',ls='--')
plt.plot(xvals,np.abs(yvalst-yvals2),color='b',ls='--')
plt.plot(xvals,np.abs(yvalst-yvals3),color='g',ls='--')

plt.xlabel('$x$')
plt.ylabel('$e^{x}$')
```

Problem 3 : 5 pts- If we list all the natural numbers below 15 that are multiples of 4 or 7, we get 4, 7, 8, 12, 14. The sum of these multiples is 45. Find the sum of all multiples of 4 or 7 below 162. Note, you will need to make use of the mod command %, which in effect finds remainders from division. Note the use below. Likewise you will also need to use the logical operation or.

```python
def modulas(max,mult1,mult2):
    temp = max -1 ##for range
    sum = 0
    for i in range(temp):
        if i % mult1 == 0:
            sum += i
        if i % mult2 == 0:
            sum += i
    return sum
```
